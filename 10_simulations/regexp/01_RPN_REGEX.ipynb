```python
@dataclass
class Token:
    value: str
    token_type: 'TokenType'

class TokenType(Enum):
    LETTER = auto()
    DIGIT = auto()
    STAR = auto()
    PIPE = auto()
```

```python
def token_generator(x: str) -> Iterable[Token]:
    for char in x:
        if char in OPERATORS:
            yield Token(char, TokenType.OPERATOR)
        elif (char in LETTERS) or (char in DIGITS):
            yield Token(char, TokenType.OPERAND)
        else:
            raise ValueError(f"Unknown character {char}")

def tokenize(x: str) -> List[Token]:
    return list(token_generator(x))
```

```python
def text_tokenize():
    assert tokenize("a") == [Token("a", TokenType.OPERAND)], "Single letter operand"
    assert tokenize("1") == [Token("1", TokenType.OPERAND)], "Single digit operand"
    assert tokenize("|") == [Token("|", TokenType.OPERATOR)], "Single operator"
    assert tokenize("*") == [Token("*", TokenType.OPERATOR)], "Single operator"
    assert tokenize("a|") == [Token("a", TokenType.OPERAND), Token("|", TokenType.OPERATOR)], "Operand followed by operator"
    assert tokenize("1*") == [Token("1", TokenType.OPERAND), Token("*", TokenType.OPERATOR)], "Operand followed by operator"
    assert tokenize("a1|*") == [Token("a", TokenType.OPERAND), Token("1", TokenType.OPERAND), Token("|", TokenType.OPERATOR), Token("*", TokenType.OPERATOR)], "Mix of operands and operators"
```

```python
def infix2rpn(expression: str) -> List[str]:
    precedence = {'|': 1, '&': 2, '*': 3}
    output = []
    stack = []

    def higher_precedence(op1, op2):
        return precedence[op1] >= precedence[op2]

    def pop_higher_precedence_operators(op):
        while stack and stack[-1].token_type == TokenType.OPERATOR and higher_precedence(stack[-1].value, op):
            output.append(stack.pop().value)

    for token in expression:
        if token.isalnum():
            output.append(token)
        elif token in '|&*':
            pop_higher_precedence_operators(token)
            stack.append(Token(token, TokenType.OPERATOR))
        elif token == '(':
            stack.append(Token(token, TokenType.OPERATOR))
        elif token == ')':
            while stack and stack[-1].value != '(':
                output.append(stack.pop().value)
            stack.pop()  # Discard the '('

    while stack:
        output.append(stack.pop().value)

    return output

def test_infix2rpn():
    assert infix2rpn('a|b') == ['a', 'b', '|']
    assert infix2rpn('(a|b)*c') == ['a', 'b', '|', 'c', '&', '*']
    assert infix2rpn('a&(b|c)*d') == ['a', 'b', 'c', '|', '&', 'd', '*', '&']

```
