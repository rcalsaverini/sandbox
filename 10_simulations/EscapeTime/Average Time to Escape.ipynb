```python
from sde import *
from tqdm import autonotebook as tqdm
```

```python
target_distance = 5
max_time = 5
dt = 0.01
sigmas = pd.Index(np.linspace(0, 10, 100), name="sigma")

stats = pd.DataFrame(index=sigmas, columns=["out", "time_to_out"])
stats.loc[:, :] = 0

for _ in tqdm.trange(100):
    for sigma in tqdm.tqdm(sigmas):
        m = Model(sigma=sigma, force=lambda t, x: 0)
        s = m.simulate_until_distance(10, 0.1, 5)
        out = s.x.iloc[-1] >= target_distance
        time_to_out = s.t.max() if out else max_time
        stats.loc[sigma] += pd.Series({"out": out, "time_to_out": time_to_out})    
```

```python
stats / 100
```
