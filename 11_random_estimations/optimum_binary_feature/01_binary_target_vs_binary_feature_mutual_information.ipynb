```python
import numpy as np
from matplotlib import pyplot as plt
```

- $D$: default, 0 or 1
- $X$: feature, 0 or 1

$$ 
MI[D, X] = H[D] - \mathbb{E}[H[D|F=f]]
$$

- Average risk: $P(D=1) = \bar{R}$
- Delta risk: 
    - $P(D=1|F=1) = R_1$
    - $P(D=1|F=0) = R_0$
- Feature coverage: $P(F=1) = C$

```python
def entropy(p):
    if p == 0.0:
        return 0.0
    if p == 1.0:
        return 0.0
    return -p * np.log(p) - (1-p) * np.log(1 - p)
```

```python
def mutual_information(conditional_risks, coverage):
    r0, r1 = conditional_risks
    r_avg = r0 * (1 - coverage) + r1 * coverage
    h0 = entropy(r0)
    h1 = entropy(r1)
    hc = coverage * h1 + (1 - coverage) * h0
    return entropy(r_avg) - hc

```

```python
cs = np.linspace(0.0, 1.0, 31)
mis = np.array([mutual_information([1.0, 0.0], c) for c in cs])
```

```python
plt.plot(cs, mis)
```

```python
mutual_information([1.0, 0.0], 0.5) / mutual_information([1.0, 0.0], 0.05)
```
