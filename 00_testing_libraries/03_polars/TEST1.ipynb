```python
import polars
import pandas
```

```python
PATH = "../01_crawl_wikipedia/dataset.parquet"
```

```python
%%timeit
pandas.read_parquet(PATH).groupby("word").agg({"ipa": list})
```

```python
%%timeit
polars.read_parquet(PATH).groupby("word")\
    .agg(polars.col("ipa").list())
```

```python
%%timeit
polars.read_parquet(PATH).lazy()\
    .groupby("word")\
    .agg(polars.col("ipa").list())
```

```python
%timeit df_pl = polars.read_parquet(PATH)
%timeit df_pd = pandas.read_parquet(PATH)
```

```python
df_pl = polars.read_parquet(PATH)
df_pd = pandas.read_parquet(PATH)

df_pl.head()
```
