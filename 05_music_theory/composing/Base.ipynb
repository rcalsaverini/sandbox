```python
from mingus.containers import NoteContainer, Bar, Track, Composition
from mingus.core import chords
from mingus.midi import midi_file_out
```

```python
from music21 import midi
```

```python


# Chord changes
chord_changes = ["Dm", "G7", "C"]
```

```python
composition = Composition()

track = Track()
for chord in chord_changes:
    bar = Bar()
    bar.place_notes(NoteContainer(chords.from_shorthand(chord)), 1)
    track.add_bar(bar)
composition.add_track(track)
```

```python
midi_file_out.write_Composition("out.mid", composition)
```

```python
mf = midi.MidiFile()
mf.open("out.mid")
mf.read()
mf.close()
s = midi.translate.midiFileToStream(mf)
s.show('midi')
```
