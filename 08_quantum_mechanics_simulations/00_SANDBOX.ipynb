```python
x_lim = (-1, 2)
num_spacial_bins = 100
num_fourier_bins = 100
```

```python
x_space = np.linspace(x_lim[0], x_lim[1], num_spacial_bins)
k_space = np.arange(-num_fourier_bins // 2, num_fourier_bins // 2)
fourier_coeffs = rnd.normal(0, 1, num_fourier_bins)
```

```python
cosines = np.cos(np.pi * x_space[:, None] * k_space[None, :])
potential = cosines @ fourier_coeffs
potential
```

```python
plt.plot(x_space, potential)
```
