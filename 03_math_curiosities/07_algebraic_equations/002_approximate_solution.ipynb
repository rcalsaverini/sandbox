```python
import numpy as np
from matplotlib import pyplot as plt

j = complex(0, 1)
```

```python
alphas = np.logspace(-10, 10, 1001)
sigmas = 1 / (1 + alphas ** 2)
```

```python
def root_magnitude(p):
    return alphas ** (1/ (p + sigmas))
    
def root_angle(p, k):
    return -np.pi * (2 * k + sigmas) / (p + sigmas)

def root(p, k):
    r = root_magnitude(p)
    theta = root_angle(p, k)
    return r * np.exp(j * theta)
```

```python
p = 10
k = 5
x = root(p, k)
```

```python
plt.plot(alphas, np.real(alphas * x ** p + x - 1))
```
