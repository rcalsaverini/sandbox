```python
import numpy as np
from matplotlib import pyplot as plt
```

$$S = \frac{1}{1+\frac{2}{1 + \frac{3}{1+\ldots}}}$$

```python
def step(x, k):
    if k == 0:
        return x
    else:
        return step(1/((2 * k - 1) + 2 * k * x), k - 1) 
```

```python
step(np.inf, 1000)
```

```python
lxs = {k: step(np.inf, k) for k in range(20)}
```

```python
xs, ys = zip(*lxs.items())
xs = np.array(xs)
ys = np.array(ys)
```

```python
# plt.xscale("log")
# plt.yscale("log")
plt.scatter(1+xs, ys)
```
