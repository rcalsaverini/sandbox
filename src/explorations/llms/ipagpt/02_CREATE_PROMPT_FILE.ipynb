```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_parquet("./dataset.parquet")
data.head()
```

```python
train_words, test_words = train_test_split(data.word.unique())
```

```python
train_data = data.loc[data.word.isin(train_words)]
test_data = data.loc[data.word.isin(test_words)]
```

```python
def prompt_from_row(row):
    if row.context_proc == "":
        return f"IPA pronunciation of '{row.word}':".replace('"', "")
    else:
        return f"IPA pronunciation of '{row.word}' (context: {row.context_proc}):".replace('"', "")
```

```python
train_prompts = [
    f'{{"prompt": "{prompt_from_row(row)}", "completion": "{row.ipa}"}}'
    for row in train_data.itertuples()
]
```

```python
test_prompts = [
    f'{{"prompt": "{prompt_from_row(row)}", "completion": "{row.ipa}"}}'
    for row in test_data.itertuples()
]
```

```python
with open("train_prompts.jsonl", "w") as file_out:
    output = "\n".join([str(x) for x in train_prompts])
    file_out.write(output)
```

```python
with open("test_prompts.jsonl", "w") as file_out:
    output = "\n".join([str(x) for x in test_prompts])
    file_out.write(output)
```
