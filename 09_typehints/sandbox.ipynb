```python
from typing import *
```

```python
class Typeclass:
    def __init__(self, klass: type):
        pass


def typeclass(klass: type) -> Typeclass:
    return Typeclass(klass)
```

```python
A = TypeVar("A")
```

```python
class Semigroup(Generic[A]):
    __mul__: Callable[[A, A], A]
```

```python
type(Semigroup)
```
